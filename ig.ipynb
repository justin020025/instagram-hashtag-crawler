```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
from selenium import webdriver
import time
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
#抓取搜尋頁面的連結，存到links
search = '返校'
browser.get('https://www.instagram.com/explore/tags/'+search+'/?hl=zh-tw')

num = 100  #num:抓取貼文的數量
if num <= 45:
    iters = 1
elif num <= 78:
    iters = 2
elif num <= 180:
    iters = 3
else:
    iters = 4+(num-180)//45

links=[]    
for i in range(iters):    
    Pagelength = browser.execute_script("window.scrollTo("+str(i)+ ",document.body.scrollHeight/"+str(i+1)+");")
    source = browser.page_source
    data=BeautifulSoup(source, 'html.parser')
    body = data.find('body')
    script = body.find('span')
    for link in script.findAll('a'): #貼文連結在 a tag
         if re.match("/p", link.get('href')):
            links.append('https://www.instagram.com'+link.get('href'))
    #sleep time is required. If you don't use this Instagram may interrupt the script and doesn't scroll through pages
    time.sleep(5) 
links = links[:num]
```

```python
links
```

```python
#至各連結抓取 hash tag, 存到stored, stored[i][j]為第 i 篇貼文的第 j 個 hash tag ,若是[],代表該篇沒hashtag
#可能要等比較久
stored = []
for link in links:
    browser.get(link)
    sub_stored = []
    tag_soup = BeautifulSoup(browser.page_source)
    tag_soup_found = tag_soup.find('div','C4VMK').find_all(name = 'a', string = re.compile('#'))
    
    for line in tag_soup_found:
        sub_stored.append(line.text[1:])
    stored.append(sub_stored)
    time.sleep(3)
```

```python
stored
```

```python
hist = dict()
for post in stored:
    for tag in post:
        if tag in hist.keys():
            hist[tag] += 1
        else:
            hist[tag] = 1
        
```

```python
hist
```

```python
len(hist) #unique hash tags
```

```python
sorted_hist = sorted(hist.items(), key = lambda x:x[1], reverse = True)
```

```python
sorted_hist
```

```python

with open('output.csv','w',encoding = 'utf-8') as f:
    for pair in sorted_hist:
        f.write(pair[0]+',')
        f.write(str(pair[1]))
        f.write('\n')
        
f.close()
```
